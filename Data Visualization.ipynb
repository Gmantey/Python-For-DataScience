```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

> Used to show the graphs of Matplot lib : %matplotlib inline

```python
x = np.linspace (0, 4 * np.pi, num = 200)
y = np.sin(x)
```

> np.linspace (0, 1, num = 200): 0 to 1 , with 200 points

```python
plt.plot(x,y);
```

#### Graph Formatting and Lables

```python
z = np.cos(x);

plt.plot(x,y,'r', label  = 'Sin(x)');
plt.plot(x,z,'b', label  = 'Cos(x)');
plt.xlim([0,6])
plt.xlabel('x');
plt.ylabel('f(x)');
plt.legend()
```

## Subplots

```python
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.plot(x,y,'r');
plt.subplot(1,2,2)
plt.plot(x,z,'b')
#plt.tight_layout;
```

```python
fig = plt.figure() #Figure is the whole canvas

axes = fig.add_axes([0.1,0.1,0.9,0.9]) # Object oriented Graphing
```

# More Types of Graphs

```python
import pandas as pd
```

```python
data = pd.read_csv('SAT.csv')
data.tail()
```

```python
sub = (data['Critical Reading Mean'] > 0)
sub = data[sub]
sub
```

```python
plt.hist(sub['School Name'])
plt.figure(figsize = (20,8))
```

### Scatter Plots

```python
plt.scatter(x = sub['Critical Reading Mean'], y = sub['Mathematics Mean'], alpha = 0.4)
plt.xlabel('Reading Score')
plt.ylabel('Math Score')
```

### 2D Histogram Plots

```python
plt.hist2d(x = sub['Critical Reading Mean'], y = sub['Mathematics Mean']);
plt.colorbar()
```

# SEABORN GRAPHING

```python
# Always use Seaborn
import seaborn as sns
import warnings
```

```python
sub.head()
```

```python
plt.figure(figsize = [10,6]) # Had to put it before it took effect
sns.distplot(sub['Number of Test Takers'] , bins = 20);
```

```python
sns.lmplot('Critical Reading Mean', 'Mathematics Mean', data = sub, aspect = 2)
# aspect = , makes the x axis 2 times larger than the y
# size = , makes the y axis larger
# hue = '', introduces another variable and tracks the way the correlation affects it
# --- Must have a correlation to work though 
# col = '', introduces another variable, breaks it down into bins.
```

```python
sns.kdeplot(sub['Critical Reading Mean'], sub['Mathematics Mean']);
# shade = True, changes the color
```
